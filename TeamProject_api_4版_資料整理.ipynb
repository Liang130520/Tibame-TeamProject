```python
df = pd.DataFrame(data_shilin, columns = data_shilin[0]) # pandas的表個型態
df = df.drop(["a", "AA12", "bn", "commid", "id", "pimg", "sq", "cp", "es", "mark", "note", "punit", "reid"], axis = 1)

# df["ans"] = data_shilin["p"] # 看要預測單價或總價
# df["ans"] = data_shilin["tp"] # 看要預測單價或總價

pd.set_option('display.max_columns', 500)
df

# 無意義資料:a(地址), AA12(無資料), bn(社區名), commid(社區ID), id(排序ID), lat(經度), lon(緯度),  pimg(照片), sq(一串不明意義文字) delete complete
# 缺失值太多, "cp", "es", "mark", "note", "punit", "reid" delete complete
```

```python
" 分割 f "
" 取得購買的第一樓層 "
def change(n1):
  if pd.isna(n1):
    return n1
  else:
    x = n1.split("/")[0].strip()
    if x.find(",") != -1:
      return x.split(",")[0].strip()
    else:
      return x

df["flr"] = df["f"].apply(change)

" 額外的第二樓層 "
df["flr"] = df["f"].apply(change)
df_flr = df["flr"].unique()

def change(n2):
  if pd.isna(n2):
    return n2
  else:
    x = n2.split("/")[0].strip()
    if (x.find(",") != -1):
      return x.split(",")[1]
    else:
      return None

df["extraflr"] = df["f"].apply(change)

" 額外的第三樓層 "
def change(n3):
  if pd.isna(n3):
    return n3
  else:
    x = n3.split("/")[0].strip()
    if (x.find(",") != -1):
      if len(x.split(",")) >= 3:
        return x.split(",")[2]
      else:
        return None
    else:
      return None
df["2extraflr"] = df["f"].apply(change)

" 額外的第四樓層 "
def change(n4):
  if pd.isna(n4):
    return n4
  else:
    x = n4.split("/")[0].strip()
    if (x.find(",") != -1):
      if len(x.split(",")) >= 3:
        if len(x.split(",")) >= 4:
          return x.split(",")[3]
        else:
          return None
      else:
        return None
    else:
      return None

df["3extraflr"] = df["f"].apply(change)

" 總樓層 "
def change(s):
  if pd.isna(s):
    return s
  else:
    return s.split("/")[1].strip()
df["totalflr"] = df["f"].apply(change)
df_totalflr = df["totalflr"].unique()
df_totalflr
```

```python
" v 切割(?房?廳?衛) " # 先切?房
def change(r1):
  if pd.isna(r1):
    return r1
  else:
    return r1.split("房")[0]
df["room"] = df["v"].apply(change)

# 再切廳
def change(r2):
  if pd.isna(r2):
    return r2
  else:
    return r2.split("廳")[0].split("房")[-1]
df["office"] = df["v"].apply(change)

# 再切衛
def change(r3):
  if pd.isna(r3):
    return r3
  else:
    return r3.split("廳")[-1].split("衛")[0]
df["toilet"] = df["v"].apply(change)

# 需切割資料:v(?房?廳?衛) depart complete
```

```python
" 貨幣轉換 "
# tp & p 有 ',', 不為數值
def change(x): 
    if pd.isna(x):
        return x
    else:
        return x.replace(",","")
    
df["tp"] = df["tp"].apply(lambda x:change(x))

def change(x2): 
    if pd.isna(x2):
        return x2
    else:
        return x2.replace(",","")
    
df["p"] = df["p"].apply(change)
```

```python
" 日期轉換 "
def change(d): 
    if pd.isna(d):
        return d
    else:
        d = str(int(d.split("/")[0]) + 1911) + "-" + d.split("/")[1] + "-" + d.split("/")[2]
        return d   
df["e"] = df["e"].apply(change)
df["e"] = pd.to_datetime(df['e'], errors = 'coerce')
```

```python
" 樓層名轉換數值"
# 類別值:AA11, b, city, el, m, msg, pu, t, twn, type
# 未知定義, 先當類別:el, m

column_dict = {"--":None,"":None,None:0, '陽台':1, '平台':1, '屋頂突出物':1, '夾層':1, '電梯樓梯間':1, '騎樓':1, '陽臺':1, '地下一層':0, '地下層':0, '全':100, '一層':1, '二層':2, '三層':3, '四層':4, '五層':5, '六層':6, '七層':7, '八層':8, '九層':9, '十層':10, '十一層':11, '十二層':12, '十三層':13, '十四層':14, '十五層':15, '十六層':16, '十七層':17, '十八層':18, '十九層':19, '二十層':20, '二十一層':21, '二十二層':22, '二十三層':23, '二十四層':24, '二十五層':25, '二十六層':26, '二十七層':27, '二十八層':28, '二十九層':29, '三十層':30, '三十一層':31, '三十二層':32, '三十三層':33, '三十四層':34, '三十五層':35, '三十六層':36, '三十七層':37, '三十八層':38, '三十九層':39, '四十層':40, '四十一層':41, '四十二層':42, '四十三層':43, '四十四層':44, '四十五層':45, '四十六層':46, '四十七層':47, '四十八層':48, '四十九層':49, '五十層':50}

df = df.replace({"flr": column_dict})
df = df.replace({"extraflr": column_dict})
df = df.replace({"2extraflr": column_dict})
df = df.replace({"3extraflr": column_dict})
df = df.replace({"totalflr": column_dict})
df = df.replace({"pu": column_dict})

" 將 "" 轉換空值 "
# 數值:bs, e, g, j, k, l, lat, lon, s, flr, extraflr, 2extraflr, 3extraflr, totlaflr, room, offic, toilet
# 未知定義, 先當數值:fi, r, tunit, unit

column_sort_dict = {"":None}

df = df.replace({"bs": column_sort_dict})
df = df.replace({"e": column_sort_dict})
df = df.replace({"g": column_sort_dict})
df = df.replace({"j": column_sort_dict})
df = df.replace({"l": column_sort_dict})
df = df.replace({"fi": column_sort_dict})
df = df.replace({"room": column_sort_dict})
df = df.replace({"office": column_sort_dict})
df = df.replace({"toilet": column_sort_dict})

df = df.drop(["f", "v"], axis = 1) # 難分辨資料已分割, 去除該資料
df
```

```python
" 補類別值 "
# 自認非空值, 待補類別值 pu

# df["pu"].value_counts() # 取得 pu 欄位內值的數量
# df["pu"].value_counts().max() # 取得 pu 類別內數量最多的數量
mostidx = df["pu"].value_counts().idxmax() # 取得 pu 內最多數量的值

df["pu"] = df["pu"].fillna(mostidx) # fillna 空值欄填入資料, mostidx 為[(最多的類別)的值]
```

```python
" 補數值 "
# 自認非空值, 待補數值 bs, fi, g, room, office, toilet

df.median() 

bs_med = df["bs"].median()
df["bs"] = df["bs"].fillna(bs_med)

fi_med = df["fi"].median()
df["fi"] = df["fi"].fillna(fi_med)

g_med = df["g"].median()
df["g"] = df["g"].fillna(g_med)

totalflr_med = df["totalflr"].median()
df["totalflr"] = df["totalflr"].fillna(totalflr_med)

room_med = df["room"].median()
df["room"] = df["room"].fillna(room_med)

office_med = df["office"].median()
df["office"] = df["office"].fillna(office_med)

toilet_med = df["toilet"].median()
df["toilet"] = df["toilet"].fillna(toilet_med)
```
