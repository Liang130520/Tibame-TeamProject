```python
import pandas as pd
from urllib.request import urlretrieve
from sklearn.datasets import load_boston

# url = "http://lib.stat.cmu.edu/datasets/boston"
# urlretrieve(url, "test.csv")

" 抓整理過的資料 "
shilin_111_data = pd.read_csv("shilin_111_house.csv", encoding = "utf-8")
df = pd.DataFrame(shilin_111_data)

" 新增答案欄位 "
df["ans"] = df["tp"]

" 顯示樣貌 "
pd.set_option('display.max_columns', 500)
df
```

```python
" 如果要再去除某項用 "
# df = df.drop(["p","tp","ans"], axis = 1)
df
```

```python
" 做 One_Hot Encoding "
# 類別值:AA11, b, city, el, m, msg, pu, t, twn, type
# 未知定義, 先當類別:el, m
pd.set_option('display.max_columns', 500)
df = pd.get_dummies(df)
# df = pd.get_dummies(df, columns = ["pu"])
df
```

```python
" 設定訓練資料跟答案 "
import numpy as np
x = np.array(df.drop(['p','tp','ans'], axis = 1))
y = np.array(df["ans"])

from sklearn.model_selection import train_test_split
# [90%x, 10%x, 90%y, 10%y]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)

" 用回歸計算 "
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=4)
reg.fit(x_train, y_train)

" 決策樹呈現 "
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(50, 50))
plot_tree(reg,
     feature_names = shilin_111_data.columns,
     filled=True,
     max_depth=2)

" 模型預測情況 "
from sklearn.metrics import r2_score
pre = reg.predict(x_test)
r2_score(y_test, pre)
```
