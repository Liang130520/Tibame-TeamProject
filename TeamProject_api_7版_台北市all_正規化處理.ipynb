下方：標準化, 取極值, 並捨去

```python
" 載入程式 "
# !pip install sklearn
from sklearn.preprocessing import StandardScaler
from math import sqrt

# 做出 idx 的Series, 方便後面刪除 row
dic = pd.Series({"idx":[]}) 
```

```python
""" 土地移轉總面積平方公尺 標準化 """
# 資料標準差前置
values = df['土地移轉總面積平方公尺'].values
values = values.reshape((len(values), 1))
# 計算標準差
scaler = StandardScaler()
scaler = scaler.fit(values)
# 取平均數
avg = float(scaler.mean_)  
# 取標準差
standard = sqrt(scaler.var_)  
print(avg, standard)

" 抓取 2.5% 極值資料 "
for i in range(len(df['土地移轉總面積平方公尺'])):
  if (abs(df["土地移轉總面積平方公尺"][i] - avg) >= standard * 3):
    if i in dic["idx"]:
      pass
    else:
      dic["idx"].append(i)
  else:
    pass

" 確認 2.5% 極值資料 "
for j in dic["idx"]:  
  df[j:j+1]
```

#### Max Min 正規化處理

```python
""" Max Min 正規化處理 """
from sklearn import preprocessing
minmax = preprocessing.MinMaxScaler()
```

```python
" 總價元 Max Min 正規化處理 "
values = df['總價元'].values
values = values.reshape((len(values), 1))
df["總價元"] = minmax.fit_transform(values)
```

```python
" 單價元平方公尺 Max Min 正規化處理 "
values = df['單價元平方公尺'].values
values = values.reshape((len(values), 1))
df["單價元平方公尺"] = minmax.fit_transform(values)
```

```python
" 所得中位數 Max Min 正規化處理 "
values = df['所得中位數'].values
values = values.reshape((len(values), 1))
df["所得中位數"] = minmax.fit_transform(values)
```
