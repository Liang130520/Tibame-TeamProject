```python
""" One-Hot Encoding """
df = pd.get_dummies(df)
df = pd.get_dummies(df, columns = ["有無管理組織","電梯"])
# df
```

```python
" 取得數值類中位數 "
med = df.median()
med
""" 更改數值 column 下的 NAN 值 """
df = df.fillna(med)
```

下方：正規化

```python
" 載入程式 "
# !pip install sklearn
from sklearn.preprocessing import StandardScaler
from math import sqrt

# 做出 idx 的Series, 方便後面刪除 row
dic = pd.Series({"idx":[]}) 
```

```python
""" 主建物面積 正規化 """
# 資料標準差前置
values = df['主建物面積'].values
values = values.reshape((len(values), 1))
# 計算標準差
scaler = StandardScaler()
scaler = scaler.fit(values)
# 取平均數
avg = float(scaler.mean_)  
# 取標準差
standard = sqrt(scaler.var_)  
print(avg, standard)

" 抓取 2.5% 極值資料 "
for i in range(len(df['主建物面積'])):
  if ((df["主建物面積"][i] - avg) >= standard * 3):
    dic["idx"].append(i)
  else:
    pass

" 確認 2.5% 極值資料 "
for j in dic["idx"]:  
  df[j:j+1]
```

```python
""" 建物移轉總面積平方公尺 正規化 """
# 資料標準差前置
values = df['主建物面積'].values
values = values.reshape((len(values), 1))
# 計算標準差
scaler = StandardScaler()
scaler = scaler.fit(values)
# 取平均數
avg = float(scaler.mean_)  
# 取標準差
standard = sqrt(scaler.var_)  
print(avg, standard)

" 抓取 2.5% 極值資料 "
for i in range(len(df['主建物面積'])):
  if ((df["主建物面積"][i] - avg) >= standard * 3):
    dic["idx"].append(i)
  else:
    pass

" 確認 2.5% 極值資料 "
for j in dic["idx"]:  
  df[j:j+1]
```
